```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
url_a="https://books.toscrape.com/catalogue/page-"
url_b=".html"
df=pd.DataFrame({
    #"id":[],
    "book_name":[],
    "price":[],
    "stock":[],
    "url":[]
})
for i in range(1,51):
    url=url_a+f"{i}"+url_b
    r=requests.get(url, verify=True, timeout=10)
    s=BeautifulSoup(r.content,"html.parser")
    divs=s.find_all('article', class_='product_pod')
    for div in divs:
        quote={}
        quote["price"]=div.find('p', class_='price_color').text
        if (str(div.find('p',class_='instock availability').text).strip())=="In stock":
            quote["stock"]=1
        else:
            quote["stock"]=0
        quote["url"]=("https://books.toscrape.com/"+div.find('a').get("href"))
        quote["book_name"]=(div.find('h3').text)
        df.loc[len(df)]=quote
df.index.name="id"
df
```
